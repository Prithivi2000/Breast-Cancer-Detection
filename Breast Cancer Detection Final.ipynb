```python
from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
```

```python
cancercopy2["Class"] = encode.fit_transform(cancercopy2["Class"])
cancercopy2["Age"] = encode.fit_transform(cancercopy2["Age"])
cancercopy2["Menopause"] = encode.fit_transform(cancercopy2["Menopause"])
cancercopy2["TumorSize"] = encode.fit_transform(cancercopy2["TumorSize"])
cancercopy2["invnodes"] = encode.fit_transform(cancercopy2["invnodes"])
cancercopy2["Nodecaps"] = encode.fit_transform(cancercopy2["Nodecaps"])
cancercopy2["Breast"] = encode.fit_transform(cancercopy2["Breast"])
cancercopy2["Breastquad"] = encode.fit_transform(cancercopy2["Breastquad"])
cancercopy2["Irradiate"] = encode.fit_transform(cancercopy2["Irradiate"])
```

```python
breast_num = strat_train_set.drop("Irradiate", axis = 1)
breast_labels = strat_train_set["Irradiate"].copy()
breast_num_test = strat_test_set.drop("Irradiate", axis = 1)
breast_labels_test = strat_test_set["Irradiate"].copy()
```

```python
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
model = XGBClassifier(random_state = 5, learning_rate =1,)
# model = RandomForestClassifier()
# model = DecisionTreeClassifier()
```

```python
breast_num_arr = breast_num.to_numpy()
breast_labels_arr = breast_labels.to_numpy()
breast_num_test_arr = breast_num_test.to_numpy()
breast_labels_test_arr = breast_labels_test.to_numpy()
```

```python
model.fit(breast_num_arr, breast_labels_arr)
```

```python
# some_data = breast_num.iloc[:5]
# some_labels = breast_labels.iloc[:5]
```

```python
y_predict = model.predict(breast_num_arr)
y_predict
```

```python
list(breast_labels_arr)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(breast_labels_arr, y_predict))
```

```python
y_train_predict = model.predict(breast_num_test_arr)
print(accuracy_score(breast_labels_test_arr, y_train_predict))
```
